```python
# Graficando un array:
# En caso de usar un array tomara los valores del mismo en el eje Y y las posiciones seran los valores n X
# Graficando un ejemplo.

array = [1,2,5,7,8,3,1]

# Graficando la posicion de "1" sera "0", la posicion de "2" sera "1", la de "5" sera "3" y asi...

plt.plot(array)
plt.show() # -> Esto me muestra el grafico que arme arriba
```

```python
# Tambien se puede representar un grafico mediante dos arrays
# Ejemplo de grafico con 2 arrays

array1 = [-3,-1,0,4,7]
array2 = [1,4,6,7,8]

plt.plot(array1,array2) # el primer array sera de eje X y el segundo array el de eje Y
plt.show()
```

```python
# Definiendo los limites de los ejes para que se vea mejor (Antes de ejecutar plt.show())
# Podemos definir los limites mediante plt.axis([xmin,xmax,ymin,ymax])
# xmin: valor minimo de eje x
# xmax: valor maximo del eje x
# ymin: valor minimo de eje y
# ymax: valor maximo de eje y
# Deben estar SI O SI EN ESE ORDEN!!! y deben ser pasados como una lista!!!!

array1 = [-3,-1,0,4,7]
array2 = [1,4,6,7,8]
plt.plot(array1,array2)
# valor minimo de ejex es -3 asi que pongo -4
# valor maximo de eje x es 7 asi que pongo 8
# valor minimo de eje y es 1 asi que pongo 0
# valor maximo de eje y es 8 asi que pongo 9
plt.axis([-4,8,0,9])
plt.show()
```

```python
# Se pueden usar formulas matematicas para pasarlas como coordenadas al grafico
import numpy as np
x = np.linspace(-2,2,500)
y = x**2

# Con esas dos variables graficamos!
plt.plot(x,y)
plt.show()
```

```python
# Agregando titulo y label a los ejes y cuadricula
# plt.title(): agrega un titulo con el string que pongamos dentro
# plt.xlabel(): agregamos una leyenda al eje x.
# plt.ylabel(): agregamos una leyenda al eje y.
# plt.grid("True"): agregando esto con el valor "True" nos pondra cuadricula al grafico

# Usando el ejemplo anterior
x = np.linspace(-2,2,500)
y = x**2

plt.plot(x,y)
plt.title('funcion cuadratica')
plt.xlabel('valores de "X"')
plt.ylabel('valores de "Y"')
plt.grid("True")
plt.show()
```

```python
# Superponiendo graficos
# Usamos el ejemplo anterio con una formula agregada

x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
plt.plot(x,y,'b--',x,y2,'r') # el "b--" significa color azul y rayas, el "r" significa rojo
# le estoy diciendo que me grafique x,y, con una linea azul de rayas ('b--')
# le estoy diciendo que me grafique x,y2 con una linea roja continua ('r')

plt.title('funcion cuadratica')
plt.xlabel('valores de "X"')
plt.ylabel('valores de "Y"')
plt.grid("True")
plt.show()
```

```python
# se puede hacer lo anterior seprando los dos graficos DENTRO DE LA PROGRMACION

x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
# Separo las lineas
plt.plot(x,y,'b--') #-> le estoy diciendo que me grafique x,y, con una linea azul de rayas ('b--')
plt.plot(x,y2,'r')#-> le estoy diciendo que me grafique x,y2 con una linea roja continua ('r')

plt.title('funcion cuadratica')
plt.xlabel('valores de "X"')
plt.ylabel('valores de "Y"')
plt.grid("True")
plt.show()
```

```python
# Agregando una etiqueta o leyenda a las lineas con label= ""
# Usando el ejemplo anterior

x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
# Separo las lineas y agrego leytenda
plt.plot(x,y,'b--',label="Funcion cuadratica") #-> le estoy diciendo que me grafique x,y, con una linea azul de rayas ('b--')
plt.plot(x,y2,'r', label="Funcion lineal")#-> le estoy diciendo que me grafique x,y2 con una linea roja continua ('r')
plt.legend(loc="best") # le digo que lo ponga en el mejor lugar las etiquetas de las lineas

# Agrego titulo a grafico y ejes
plt.title('funcion cuadratica')
plt.xlabel('valores de "X"')
plt.ylabel('valores de "Y"')
plt.grid("True") # agregando esto con el valor "True" nos pondra cuadricula al grafico
plt.show()
```

```python
# Creando la misma garfica en dos por separado
# plt.subplot(x1,x2,x3)
# x1 = cantidad de filas 
# x2 = cantidad de columnas 
# x3 = posicion del subplot
# Uso del caso anterior

x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
# Separo las lineas y agrego leytenda
plt.subplot(1,2,1) # primera fila, segunda columna, primer subplot
plt.plot(x,y,'b--') 
plt.subplot(1,2,2) #primera fila, segunda columna, segundo subplot
plt.plot(x,y2,'r',)

plt.show()
```

```python
# redimensionando los graficos anteriores con plt.figure(figsize=x1,x2)
# x1 = valor de ancho del grafico
# x2 = valor de alto del grafico
# uso de ejemplo anterior
# fontsize=x le da tamao a los titulos o labels, donde x es un numero

x = np.linspace(-2,2,500)
y = x**2
y2 = x+1

# Redimensionando grafico
plt.figure(figsize=(15,6))
# Separo las lineas y agrego leytenda
plt.subplot(1,2,1) # primera fila, segunda columna, primer subplot
plt.plot(x,y,'b--') 
plt.title('Grafico de una funcion', fontsize=20)
plt.xlabel("eje X para funcion",fontsize=14)
plt.subplot(1,2,2) #primera fila, segunda columna, segundo subplot
plt.title('Grafico de una recta', fontsize=20)
plt.plot(x,y2,'r',)
plt.xlabel("eje X para recta",fontsize=14)

plt.show()
```

```python
# Graficnado Histogramas
# usando plt.hist(data,bins=x,rwidth=)
# data: es el conjunto de datos a graficar
# bins: cantidad de columnas a graficar de esos datos
# rwidth: ancho de las columnas graficadas

data = [1,1.8,2,2.1,2.3,3,3.1,3.3,3.3,3.6]

# combinando con subplots
plt.subplot(2,1,1)
plt.hist(data, bins=10,rwidth=0.8)

plt.subplot(2,1,2)
plt.hist(data, bins=[1,1.5,2,2.5,3],rwidth=0.95)
plt.xlabel('Valores',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.show()
```

```python
# Aplicando graficos a un caso con datos de salarios de data scientist, frontend y backend
# Creando las listas de salarios para cada caso y la lista de ejex para edades

edad_x = [20,25,30,35,40,45,50,55,60,65]
salario_cd_y = [20000,30000,33000,40000,42000,45000,47000,60000,70000,80000] # DATA SCIENTIST
salario_fe_y = [10000,15000,17000,27000,30000,32000,35000,40000,47000,50000] # FRONTEND
salario_be_y = [15000,20000,25000,27000,35000,40000,45000,50000,55000,60000] #BACKEND
```

```python
# Grafico simple de salario de data scientist con respecto a e su edad
# utilzando pyplot

plt.plot(edad_x,salario_cd_y)
plt.show()
```

```python
# Agregando mas cosas al grafico anterior como titulo y leyenda de los ejes

plt.plot(edad_x,salario_cd_y)
plt.title('Salario medio de un data scientist en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y
plt.grid(True) # el grid es opcional
plt.show()
print("Nota: se agrega 'fontsize=' en los diferentes titulos para agrandar las letras")
```

```python
# Agregando los datos de backend y de frontend a mi grafico
# Tambien se agregan 'labels' para cada linea NOTA! SI SE AGREGA LABELS HAY QUE PONER LEGEND!

plt.plot(edad_x,salario_cd_y, label='data scientist') # corresponde a linea de data scientist
plt.plot(edad_x,salario_fe_y, label='Front-end') # corresponde a linea de Front-end
plt.plot(edad_x,salario_be_y, label='Back-end') # corresponde a linea de Back-end

plt.title('Salario medio en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y

plt.legend() # confirmo de agregar los labels 
plt.grid(True) # el grid es opcional
plt.show()
```

```python
# Agregando nuevas propiedades a las lineas graficadas
# linewidth = es el ancho de la linea,toma como parametro un numero entero
# marker = "." son las marcas donde la linea toma valores representativos, en este caso se marca con un punto
# markersize = es el tamao del marker, toma como parametro un numero entero
# linestyle = "--" es el estilo de linea graficada, en este caso seran lineas de puntos
# color = en este caso es el color de linea, pudiendo usar colores o sus codificaciones
# viendo todo esto en un ejemplo de la linea de data scientist

plt.plot(edad_x,salario_cd_y, label='data scientist',linewidth=3,marker=".",markersize=15,color='red',linestyle="--") 
# corresponde a linea de data scientist
plt.plot(edad_x,salario_fe_y, label='Front-end') # corresponde a linea de Front-end
plt.plot(edad_x,salario_be_y, label='Back-end') # corresponde a linea de Back-end

plt.title('Salario medio en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y

plt.legend() # confirmo de agregar los labels 
plt.grid(True) # el grid es opcional
plt.show()
```

```python
# Creando linea de salario medio, debo plotear una linea que tenga el promedio de los 3 salarios
# si hago operaciones matematicas importo numpy para sumar los arrays y despues dividirlos

# IMPORTANDO NUMPY PARA OPERAR CON ARRAYS
import numpy as pd

plt.plot(edad_x,salario_cd_y, label='Data Scientist') 
# corresponde a linea de data scientist
plt.plot(edad_x,salario_fe_y, label='Front-end') # corresponde a linea de Front-end
plt.plot(edad_x,salario_be_y, label='Back-end') # corresponde a linea de Back-end

# CREACION DE LA LINEA DE SALARIO MEDIO EN BASE A LOS SALARIOS DE ARRIBA
sal_med = np.sum([salario_be_y,salario_cd_y,salario_fe_y], axis=0)/3
plt.plot(edad_x,sal_med,linewidth=2,color='black',linestyle="--",label='Salario medio')

plt.title('Salario medio en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y

plt.legend() # confirmo de agregar los labels 
plt.grid(True) # el grid es opcional
plt.show()
```

```python
# Graficando y ploteando entre dos valores con plt.fill_between(ejex, ejeya, ejeyb)
# ejex hace referencia al eje x
# ejeya hace referencia al primer valor de Y
# ejetb hace referencia al segundo valor de Y
# para este ejemplo uso los valores de back-end y el salario promedio
# usando los valores que ya venimos trabajando


plt.plot(edad_x,salario_be_y, label='Back-end') # corresponde a linea de Back-end
sal_med = np.sum([salario_be_y,salario_cd_y,salario_fe_y], axis=0)/3
plt.plot(edad_x,sal_med,linewidth=2,color='black',linestyle="--",label='Salario medio')

# Creando grafico de ploteo entre back-end y salario medio
plt.fill_between(edad_x,salario_be_y,sal_med)

plt.title('Salario medio en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y

plt.legend() # confirmo de agregar los labels 
plt.grid(True) # el grid es opcional
plt.show()
```

```python
# Mejorando el grafico de ploteo entre dos valores con algunos argumentos
# alpha=0.33 este valor cuanto mas cercano a 1 mas oscuro va a ser
# color = 'green' para este caso quiero que pinte en verde los valores entre esas dos lineas
# usando nuevamente el grafico anterior

plt.plot(edad_x,salario_be_y, label='Back-end') # corresponde a linea de Back-end
sal_med = np.sum([salario_be_y,salario_cd_y,salario_fe_y], axis=0)/3
plt.plot(edad_x,sal_med,linewidth=2,color='black',linestyle="--",label='Salario medio')

# Creando grafico de ploteo entre back-end y salario medio (Y AGREGANDO COSAS NUEVAS!)
plt.fill_between(edad_x,salario_be_y,sal_med, color='green',interpolate=True
                 alpha=0.25, label='Valores intermedios')

plt.title('Salario medio en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y

plt.legend() # confirmo de agregar los labels 
plt.grid(True) # el grid es opcional
plt.show()
```

```python
# podemos decirle que pinte de un color lo que estan por encima y de otro color lo que estan por debajo
# usamos where para darle un condicional, por ejemplo donde sal_med > salario_be_y
# llamando al ejemplo anterior pero con separacion de colores segun where

plt.plot(edad_x,salario_be_y, label='Back-end') # corresponde a linea de Back-end
sal_med = np.sum([salario_be_y,salario_cd_y,salario_fe_y], axis=0)/3
plt.plot(edad_x,sal_med,linewidth=2,color='black',linestyle="--",label='Salario medio')

# Creando grafico de ploteo entre back-end y salario medio (Y AGREGANDO COSAS NUEVAS!)
plt.fill_between(edad_x,salario_be_y,sal_med, color='yellow',interpolate=True,
                 alpha=0.25, label='backend menor la media',where = sal_med>salario_be_y)
plt.fill_between(edad_x,salario_be_y,sal_med, color='blue',interpolate=True,
                 alpha=0.25, label='backend mayor a la media',where = sal_med<salario_be_y)

plt.title('Salario medio en USD', fontsize =15) #agregando titulo
plt.xlabel('edad', fontsize=10) # agregando leyenda en eje x
plt.ylabel('salario', fontsize=10) # agregando leyenda en eje y

plt.legend() # confirmo de agregar los labels 
plt.grid(True) # el grid es opcional
plt.show()
```
